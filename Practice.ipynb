```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
def construct_bst_from_postorder(arr):
    root = Node(arr[-1])
    stack = [root]
    for i in range(len(arr) - 2, -1, -1):
        node = Node(arr[i])
        last_node = stack[-1]
        while stack and stack[-1] < node.data:
            last_node = stack.pop()
            
        if last_node.data > node.data
            last_node.left = node
        else:
            last_node.right = node
        stack.append(node)
        
def is_same(T, S):
    if S.left is None and S.right is None:
        return True

    if S.left is None or S.right is None:
        return False
    
    return S.data == T.data and (S.left.data == T.left.data and S.right.data == T.right.data)

def string_permute(str, start, end):
    if start == end:
        print("".join(str))
    else:
        for i in range(start, end + 1):
            str[start], str[end] = str[end], str[start]
            permute(str, start + 1, end)
            str[start], str[end] = str[end], str[start]
            start += 1
            
def binomial_coeff(n, k):
    res = 1
    if k > n-k:
        k = n-k
        
    for i in range(k):
        res = res * (n - i)
        res = res / (i + 1)
    return res



def smallest_window(string, text):
    if len(str) < len(text):
        return -1
    hash_text = [0] * 256
    hash_string = [0] * 256
    
    for char in text:
        hash_text[ord(char)] += 1
        
    count = 0
    start = 0
    start_ind = -1
    smallest_window = len(string)
    
    for i in range(len(string)):
        hash_string[ord(string[i])] += 1
        
        if hash_string[ord(string[i])] <= hash_text[ord(string[i])] and hash_text[ord(string[i])] > 0:
            count += 1
            
        if count == len(text):
            while hash_string[ord(string[start])] > hash_text[ord(string[start])] or hash_text[ord(string[start])] == 0:
                if hash_string[ord(string[start])] > hash_text[ord(string[start])]:
                    hash_string[ord(string[start])] -= 1
                start += 1
                
            current_window_size = i - start + 1
            if smallest_window > current_window_size:
                smallest_window = current_window_size
                start_ind = start
                
    if start_ind == -1:
        return -1
    else:
        return string[start_ind:(start_ind + smallest_window)]
                
```
